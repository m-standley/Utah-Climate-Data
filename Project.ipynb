## Join, Definition, Filter - OUTDATED

```python
OBdf.shape[0]
```

```python
codeNamesdf.shape[0]
```

```python
# Merging the two databases 'OBdf', which has observed values, and 'codeNamesdf', which has station names and was derived from the text file CodeNames.txt
finaldf = pd.merge(OBdf, codeNamesdf, left_on = 'station', right_on = 'Code', how = 'left')
```

```python
finaldf.head(20)
```

```python
finaldf.shape[0]
```

```python
# Making a dictionary called 'meaning' for datatype definitions
meaning = {
    'PRCP' : 'Precipitation',
    'SNOW' : 'Snowfall',
    'SNWD' : 'Snow depth',
    'TMAX' : 'Maximum Temperature',
    'TMIN' : 'Minimum Temperature',
    'AWND' : 'Average Daily Cloudiness',
    'EVAP' : 'Evaporation',
    'WESD' : 'Percent of Possible Sunlight',
    'SN*#' : 'Minimum Soil Temperature',
    'SX*#' : 'Maximum Soil Temperature'
}
```

```python
finaldf.replace(meaning, regex = True, inplace = True)
```

```python
finaldf.drop(["attributes", "Code"], axis = 1, inplace = True)
```

```python
finaldf.head(20)
```

```python
filtered_df = finaldf[finaldf['datatype'] == 'Precipitation']
filtered_df.head()
```
